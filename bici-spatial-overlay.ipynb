```python
%matplotlib inline

import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 15)
```

```python
graticule = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_graticules_15.geojson')
countries = gpd.read_file('https://d2ad6b4ur7yvpq.cloudfront.net/naturalearth-3.3.0/ne_50m_admin_0_countries.geojson')
bicycles = gpd.read_file('https://www.chapelhillopendata.org/api/v2/catalog/datasets/bicycle-crash-data-chapel-hill-region/exports/geojson')
```

```python
print(graticule.crs, countries.crs, bicycles.crs)
```

```python
fig, ax = plt.subplots()

graticule.plot(ax=ax, color='#295896', linewidth=.4, zorder=1);
countries.plot(ax=ax, edgecolor='white', color='#cccccc', zorder=2);
bicycles.plot(ax=ax, color='#fd6a05', zorder=3, markersize=8);

ax.set(title="World map projected in WGS84 with bicycle crashes", 
       xlabel="X coordinates in degrees",
       ylabel="Y coordinates in degrees")
```
